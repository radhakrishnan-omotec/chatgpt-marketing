**Age Range:**
- 17 -28  ( Range 1)
- 28 - 39 ( Range 2)
- 40 - 50 ( Range 3)
- .>50    ( Range 4)

In this modified code, we first read the CSV file into a DataFrame using pd.read_csv().

We then define the age ranges as ['0-28', '29-50', '>50'] and the corresponding labels as [0, 1, 2].

Next, we create a dictionary age_mapping that maps the age ranges to the corresponding labels. The zip function is used to pair the age_ranges list with age_labels, and then the dict() constructor is used to create the mapping dictionary.

We use the map() function to apply the age_mapping to the 'age_range' column, which will convert the age ranges into the corresponding age labels.

The result will be a new column 'Age Label' in the DataFrame, containing the numerical age labels (0, 1, 2) for each age range based on the mapping.

```python
# Define the age ranges and labels
age_ranges = ["17-28", "28-39", "40-50", "> 50",]
age_labels = [0, 1, 2, 3]

# Map age_range values to age_labels
age_mapping = dict(zip(age_ranges, age_labels))
print(age_mapping)
print()
print(input_df["Age "])
input_df['Age_New'] = input_df['Age '].map(age_mapping)

# Display the DataFrame with age labels
print(input_df['Age_New'])
```

## F. Categorical plots:

```python
# CATEGORICAL PLOTS PYTHON CODE SAMPLE


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read "AI_Responses.CSV File"
data =  pd.read_csv("/content/test-repo/AI_Reponses.csv")

data.head(1)


# Create a DataFrame from the data
df = pd.DataFrame(data)

# Set up the figure and axes
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Create categorical plots using Seaborn

# Example 1: Count plot for a categorical variable
plt.subplot(2, 2, 1)
sns.countplot(data=df, x='Title_new')  # DEA Correction to the appropriate name
plt.xticks(rotation=90)
plt.title("Count Plot of Current Job Position")         # DEA Correction to the appropriate name

# Example 2: Box plot for a categorical variable vs a numerical variable
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='Title_new', y='Age_New')    # DEA Correction to the appropriate name
plt.xticks(rotation=90)
plt.title("Box Plot of Age by Job Position")                     # DEA Correction to the appropriate name

# Example 3: Violin plot for a categorical variable vs a numerical variable
plt.subplot(2, 2, 3)
sns.violinplot(data=df, x='Title_new', y='Age_New') # DEA Correction to the appropriate name
plt.xticks(rotation=90)
plt.title("Violin Plot of Age by Job Position")                  # DEA Correction to the appropriate name

# Example 4: Bar plot for the distribution of responses to a question
plt.subplot(2, 2, 4)
sns.barplot(data=df.iloc[:, 4:14], ci=None)                      # DEA Correction to the appropriate name
plt.title("Bar Plot of Responses to Q1-Q10")                     # DEA Correction to the appropriate name
plt.xticks(rotation=90)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
