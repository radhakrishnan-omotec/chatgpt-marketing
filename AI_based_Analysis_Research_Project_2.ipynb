**Age Range:**
- 17 -28  ( Range 1)
- 28 - 50 ( Range 2)
- .>50     ( Range 3)

In this modified code, we first read the CSV file into a DataFrame using pd.read_csv().

We then define the age ranges as ['0-28', '29-50', '>50'] and the corresponding labels as [0, 1, 2].

Next, we create a dictionary age_mapping that maps the age ranges to the corresponding labels. The zip function is used to pair the age_ranges list with age_labels, and then the dict() constructor is used to create the mapping dictionary.

We use the map() function to apply the age_mapping to the 'age_range' column, which will convert the age ranges into the corresponding age labels.

The result will be a new column 'Age Label' in the DataFrame, containing the numerical age labels (0, 1, 2) for each age range based on the mapping.

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/test-repo/Test Responses 3.csv')

# Define the age ranges and labels
age_ranges = ["17-28", "28-39", "40-50", ">50",]
age_labels = [1, 2, 3, 4]

# Map age_range values to age_labels
age_mapping = dict(zip(age_ranges, age_labels))
print(age_mapping)
print()
print(df["Age "])
df['Age Label'] = df['Age '].map(age_mapping)

# Display the DataFrame with age labels
print(df['Age Label'])
```
